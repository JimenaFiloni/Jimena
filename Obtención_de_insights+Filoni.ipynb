El dataset sobre la evaluación de jugadores de baloncesto de la NBA, se selecciono las siguientes variables mencionadas:


- player_name (nombre del jugador): El nombre del jugador de baloncesto.

- team_abbreviation (abreviatura del equipo): La abreviatura o código que representa al equipo al que pertenece el jugador.

- age (edad): La edad del jugador.

- player_height (altura del jugador): La altura del jugador, generalmente en centímetros o pies.

- player_weight (peso del jugador): El peso del jugador, generalmente en kilogramos o libras.

- college (universidad): La universidad de la que proviene el jugador, en caso de que haya jugado a nivel universitario.

- country (país): El país de origen del jugador.

- draft_year (año del diagrama): El año en que tuvo lugar el evento de selección (posiblemente el draft).

- draft_round (ronda del diagrama): La ronda en la que fue seleccionado el jugador en el evento de selección.

- draft_number (número_del_diagrama): El número de selección del jugador en el evento de selección.

- pts (puntos): Puntos promedio por partido del jugador.

- reb (rebotes): Rebotes promedio por partido del jugador.

- ast (asistencias): Asistencias promedio por partido del jugador.

- net_rating: Clasificación neta del jugador, que puede estar relacionada con el rendimiento general del equipo mientras el jugador está en la cancha.

- oreb_pct: Porcentaje de rebotes ofensivos.

- dreb_pct: Porcentaje de rebotes defensivos.

- usg_pct (porcentaje de uso): Porcentaje de uso que mide la proporción de jugadas del equipo que involucran al jugador.

- ts_pct (porcentaje de tiros): Porcentaje de efectividad de tiros verdaderos.

- ast_pct (porcentaje de asistencias): Porcentaje de asistencias que mide la proporción de canastas de un jugador que son asistencias.

- temp (temporada): La temporada en la que se recopilaron estos datos.


Estas variables proporcionan información detallada sobre la carrera y el rendimiento de los jugadores de baloncesto en un contexto específico, y parecen ser útiles para analizar y evaluar su contribución al equipo.

```python
# Asegurarnos de que las columnas tienen los tipos de datos correctos
df_seasons['age'] = pd.to_numeric(df_seasons['age'], errors='coerce')
df_seasons['player_height'] = pd.to_numeric(df_seasons['player_height'], errors='coerce')
df_seasons['player_weight'] = pd.to_numeric(df_seasons['player_weight'], errors='coerce')
df_seasons['gp'] = pd.to_numeric(df_seasons['gp'], errors='coerce')
df_seasons['pts'] = pd.to_numeric(df_seasons['pts'], errors='coerce')
df_seasons['reb'] = pd.to_numeric(df_seasons['reb'], errors='coerce')
df_seasons['ast'] = pd.to_numeric(df_seasons['ast'], errors='coerce')
df_seasons['net_rating'] = pd.to_numeric(df_seasons['net_rating'], errors='coerce')
df_seasons['oreb_pct'] = pd.to_numeric(df_seasons['oreb_pct'], errors='coerce')
df_seasons['dreb_pct'] = pd.to_numeric(df_seasons['dreb_pct'], errors='coerce')
df_seasons['usg_pct'] = pd.to_numeric(df_seasons['usg_pct'], errors='coerce')
df_seasons['ts_pct'] = pd.to_numeric(df_seasons['ts_pct'], errors='coerce')
df_seasons['ast_pct'] = pd.to_numeric(df_seasons['ast_pct'], errors='coerce')
df_seasons = df_seasons.dropna(subset=['age', 'player_height', 'player_weight', 'gp', 'pts', 'reb', 'ast'])
```

#Análisis Exploratorio y Visualización
##Gráficos y Conclusiones

```python
#Distribución de la Edad de los Jugadores:
plt.figure(figsize=(10, 6))
sns.histplot(df_seasons['age'], bins=15, kde=True, color='skyblue')
plt.title('Distribución de la Edad de los Jugadores de la NBA')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

Conclusión:
La mayoría de los jugadores de la NBA están en el rango de edad de 20 a 30 años, lo que indica que la liga está dominada por jugadores jóvenes que están en su pico físico y atlético.

```python
#Relación entre Altura y Puntos Promedio
plt.figure(figsize=(10, 6))
sns.scatterplot(x='player_height', y='pts', data=df_seasons, hue='age', palette='viridis')
plt.title('Relación entre Altura y Puntos Promedio')
plt.xlabel('Altura (cm)')
plt.ylabel('Puntos Promedio')
plt.show()
```

Conclusión:
No hay una correlación clara entre la altura y el promedio de puntos, lo que sugiere que el éxito en anotación no está estrictamente relacionado con la altura, sino posiblemente con habilidades y roles específicos en el equipo.

```python
#Relación entre Peso y Rebotes Promedio
plt.figure(figsize=(10, 6))
sns.scatterplot(x='player_weight', y='reb', data=df_seasons, hue='age', palette='coolwarm')
plt.title('Relación entre Peso y Rebotes Promedio')
plt.xlabel('Peso (kg)')
plt.ylabel('Rebotes Promedio')
plt.show()
```

Conclusión:
Hay una tendencia leve que muestra que jugadores con más peso tienden a tener más rebotes, lo que puede indicar que los jugadores más pesados ocupan posiciones que requieren mayor presencia física bajo el aro.

```python
#Impacto del Net Rating en el Uso Porcentual
plt.figure(figsize=(10, 6))
sns.scatterplot(x='net_rating', y='usg_pct', data=df_seasons, hue='age', palette='plasma')
plt.title('Impacto del Net Rating en el Uso Porcentual')
plt.xlabel('Net Rating')
plt.ylabel('Uso Porcentual')
plt.show()
```

Conclusión:
Los jugadores con un alto uso porcentual (usg_pct) tienden a tener un net rating más variado. Esto sugiere que los jugadores que manejan más el balón y tienen mayor participación en el juego pueden tener un impacto significativo tanto positivo como negativo en el equipo.

```python
#Distribución del TS% (True Shooting Percentage)
plt.figure(figsize=(10, 6))
sns.histplot(df_seasons['ts_pct'], bins=15, kde=True, color='lightgreen')
plt.title('Distribución del True Shooting Percentage (TS%)')
plt.xlabel('TS%')
plt.ylabel('Frecuencia')
plt.show()
```

Conclusión:
La distribución del TS% muestra que la mayoría de los jugadores tienen un porcentaje de tiro verdadero en el rango del 50% al 60%, indicando una eficiencia razonable en sus intentos de anotación.

#Análisis complementario

```python
#Análisis de Clusters y Segmentación de Jugadores
# Selección de características para clustering
features = df_seasons[['player_height', 'player_weight', 'age', 'pts', 'reb', 'ast']]

# Normalización de datos
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Aplicación de KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
df_seasons['cluster'] = kmeans.fit_predict(scaled_features)

# Visualización de clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='player_height', y='pts', hue='cluster', data=df_seasons, palette='viridis')
plt.title('Clusterización de Jugadores por Altura y Puntos Promedio')
plt.xlabel('Altura (cm)')
plt.ylabel('Puntos Promedio')
plt.legend(title='Cluster')
plt.show()
```

Conclusión:
La clusterización puede revelar grupos de jugadores con características similares, como "anotadores altos", "reboteadores pesados" o "jugadores jóvenes y prometedores". Esto puede ser útil para identificar patrones y estrategias de juego.

```python
#Análisis de Rendimiento por Equipos y Coaches
team_performance = df_seasons.groupby('team_abbreviation')[['pts', 'reb', 'ast']].mean().reset_index()

plt.figure(figsize=(14, 8))
sns.barplot(x='team_abbreviation', y='pts', data=team_performance, palette='muted')
plt.title('Puntos Promedio por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Puntos Promedio')
plt.xticks(rotation=90)
plt.show()
```

**Explicación del grafico de rendimiento por equipo**
Agrupamos el dataset por la abreviatura del equipo (team_abbreviation) y calculamos el promedio de las columnas pts, reb y ast. Luego, restablecemos el índice para obtener un DataFrame plano.
Y se crea un gráfico de barras que muestra los puntos promedio por equipo.

Gráficos y Conclusiones Adicionales
Podemos extender el análisis del rendimiento por equipos con más gráficos y conclusiones sobre otras estadísticas clave como rebotes y asistencias.

```python
#Gráfico de Barras para Rebotes Promedio por Equipo:
plt.figure(figsize=(14, 8))
sns.barplot(x='team_abbreviation', y='reb', data=team_performance, palette='muted')
plt.title('Rebotes Promedio por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Rebotes Promedio')
plt.xticks(rotation=90)
plt.show()
```

```python
#Gráfico de Barras para Asistencias Promedio por Equipo
plt.figure(figsize=(14, 8))
sns.barplot(x='team_abbreviation', y='ast', data=team_performance, palette='muted')
plt.title('Asistencias Promedio por Equipo')
plt.xlabel('Equipo')
plt.ylabel('Asistencias Promedio')
plt.xticks(rotation=90)
plt.show()
```
